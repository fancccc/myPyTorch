```python
col_object = list(df.select_dtypes('object').columns)
col_object
```

```python
#类别变量众数填充
df[col_object] = df[col_object].fillna(df[col_object].mode())
```

```python
col_numerical = list(filter(lambda x:x not in col_object, list(df.columns)))
col_numerical
```

```python
#数值变量均值填充
df[col_numerical] = df[col_numerical].fillna(df[col_numerical].mean())
```

```python
df['issueDate'] = df['issueDate'].apply(lambda x:pd.to_datetime(x,format='%Y-%m-%d'))
df['issueDate'] 
```

```python
nowtime = datetime.datetime.strptime('2022-08-01','%Y-%m-%d')
df['issueDateDT'] = df['issueDate'].apply(lambda x: nowtime - x).dt.days
df['issueDateDT']
```

```python
df['earliesCreditLine'] = df['earliesCreditLine'].apply(lambda x:2022 - int(re.findall('\d+',x)[0]))
df['earliesCreditLine']
```

```python
#df['employmentLength'] = pd.cut(df['employmentLength'].apply(lambda x:int(re.findall('\d+', str(x)+'0')[0])),[0,2,5,10],labels=[0,1,2])
df['employmentLength'] = df['employmentLength'].apply(lambda x:float(((re.findall('\d+', str(x)))+[np.nan])[0]))
df['employmentLength'] = df['employmentLength'].fillna(df['employmentLength'].median())
```

```python
# 部分类别特征
cate_features = ['grade', 'subGrade', 'employmentTitle', 'homeOwnership', 'verificationStatus',
                'purpose', 'postCode', 'regionCode', 'initialListStatus', 'applicationType',
                'title', 'term']
for f in cate_features:
    print(f, '类型数：', df[f].nunique())
```

```python
col_object = ['grade', 'subGrade']
for col in col_object:
    df[col] = df[col].factorize(sort=True)[0]
```

```python
numerical_features = list(filter(lambda i:i not in cate_features + ['id', 'isDefault', 'issueDateDT'], df.columns))
numerical_features
for f in numerical_features:
    print(f, '类型数：', df[f].nunique())
```

```python
numerical_features.remove('term')
numerical_features.remove('policyCode')
```

```python
def find_outliers_by_3segama(data,fea):
    data_std = np.std(data[fea])
    data_mean = np.mean(data[fea])
    outliers_cut_off = data_std * 3
    lower_rule = data_mean - outliers_cut_off
    upper_rule = data_mean + outliers_cut_off
    data[fea+'_outliers'] = data[fea].apply(lambda x:str('异常值') if x > upper_rule or x < lower_rule else '正常值')
    return data
for fea in numerical_features:
    df = find_outliers_by_3segama(df,fea)
```

```python
distinct_fea = []
for fea in numerical_features:
    print(df[fea+'_outliers'].value_counts()/800000)
    g = df.groupby(fea+'_outliers')['isDefault'].sum()
    v = df[fea+'_outliers'].value_counts()
    print('正常值违约率：',g['正常值']/v['正常值'])
    try:
        print('异常值违约率：',g['异常值']/v['异常值'])
        if abs(g['正常值']/v['正常值'] - g['异常值']/v['异常值']) >= 0.20:
            distinct_fea.append(fea+'_outliers')
    except:
        print('异常值违约率：0')
    print('*'*10)
```

```python
distinct_fea
for col in distinct_fea:
    df[col] = df[col].map({'正常值':0, '异常值':1})
```
